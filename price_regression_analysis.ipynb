```python
data = pd.get_dummies(data)
```

```python
data.head()
```

```python
X = data.drop('Price', axis=1).values
y = data['Price'].values
```

```python
# Splitting dataset into train and test set
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

```python
print("Shape of X_train: ",X_train.shape)
print("Shape of X_test: ", X_test.shape)
print("Shape of y_train: ",y_train.shape)
print("Shape of y_test",y_test.shape)
```

# Regression Models

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(X_train, y_train)
```

```python
from sklearn.model_selection import cross_val_score

# Predicting Cross Validation Score the Test set results
cv_linear = cross_val_score(estimator=lr, X=X_train, y=y_train, cv=10)
print("CV: ", round(cv_linear.mean(),2))
```

```python
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error

#Predicting r2 score on train set
y_train_pred = lr.predict(X_train)
train_score = r2_score(y_train_pred, y_train)

#Predicting r2 score on test set
y_pred = lr.predict(X_test)
test_score = r2_score(y_pred, y_test)

# Predicting RMSE on test result
rmse_linear = np.sqrt(mean_squared_error(y_test, y_pred))

print("CV: ", round(cv_linear.mean(),2))
print("Train Score: ", round(train_score,2))
print("Test Score: ", round(test_score,2))
print("Test Score(RMSE): ", round(rmse_linear,2))
```

```python
## Polynomial Regression Degree 2
from sklearn.preprocessing import PolynomialFeatures
poly_reg = PolynomialFeatures(degree=2)
X_poly = poly_reg.fit_transform(X_train)
poly_reg.fit(X_poly, y_train)

regression_poly2 = LinearRegression()
regression_poly2.fit(X_poly, y_train)
```

```python
cv_linear_poly2 = cross_val_score(estimator=regression_poly2, X=X_train, y=y_train, cv=10)

#Predicting r2 score on train set
y_train_pred_poly2 = regression_poly2.predict(poly_reg.fit_transform(X_train))
r2_score_polyreg_train = r2_score(y_train, y_train_pred_poly2)

#Predicting r2 score on test set
y_pred_poly2 = regression_poly2.predict(poly_reg.fit_transform(X_test))
r2_score_polyreg_test = r2_score(y_test, y_pred_poly2)

# Predicting RMSE on test result
rmse_linear_poly = (np.sqrt(mean_squared_error(y_test, y_pred_poly2)))

print("CV: ", round(cv_linear_poly2.mean(),2))
print("Train Score: ", round(r2_score_polyreg_train,2))
print("Test Score: ", round(r2_score_polyreg_test,2))
print("Test Score(RMSE): ", rmse_linear_poly)
```

## Ridge Regression

```python
from sklearn.linear_model import Ridge
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import PolynomialFeatures

steps = [('scaler', StandardScaler()),
         ('poly', PolynomialFeatures(degree=3)),
         ('model', Ridge(alpha=1777, fit_intercept=True))
        ]

ridge_pipe = Pipeline(steps)
ridge_pipe.fit(X_train, y_train)
```

```python
cv_ridge = cross_val_score(estimator=ridge_pipe, X=X_train, y=y_train, cv=10)

#Predicting r2 score on train set
y_train_ridge = ridge_pipe.predict(X_train)
r2_score_ridge_train = r2_score(y_train, y_train_ridge)

#Predicting r2 score on test set
y_pred_ridge = ridge_pipe.predict(X_test)
r2_score_ridge_test = r2_score(y_test, y_pred_ridge)

# Predicting RMSE on test result
rmse_ridge = (np.sqrt(mean_squared_error(y_test, y_pred_ridge)))

print("CV: ", round(cv_ridge.mean(),2))
print("Train Score: ", round(r2_score_ridge_train,2))
print("Test Score: ", round(r2_score_ridge_test,2))
print("Test Score(RMSE): ", rmse_ridge)
```

# Lasso Regression

```python
from sklearn.linear_model import Lasso
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import PolynomialFeatures

steps = [('scaler', StandardScaler()),
         ('poly', PolynomialFeatures(degree=3)),
         ('model', Lasso(alpha=2.36, fit_intercept=True, tol = 0.0199, max_iter=2000))
        ]

lasso_pipe = Pipeline(steps)
lasso_pipe.fit(X_train, y_train)
```

```python
cv_lasso = cross_val_score(estimator=lasso_pipe, X=X_train, y=y_train, cv=10)

#Predicting r2 score on train set
y_train_lasso = lasso_pipe.predict(X_train)
r2_score_lasso_train = r2_score(y_train, y_train_lasso)

#Predicting r2 score on test set
y_pred_lasso = lasso_pipe.predict(X_test)
r2_score_lasso_test = r2_score(y_test, y_pred_lasso)

# Predicting RMSE on test result
rmse_lasso = (np.sqrt(mean_squared_error(y_test, y_pred_lasso)))

print("CV: ", round(cv_lasso.mean(),2))
print("Train Score: ", round(r2_score_lasso_train,2))
print("Test Score: ", round(r2_score_lasso_test,2))
print("Test Score(RMSE): ", rmse_lasso)
```

# Suport Vector Regression

```python
# Feature Scaling
from sklearn.preprocessing import StandardScaler
sc_X = StandardScaler()
sc_y = StandardScaler()

X_scaled = sc_X.fit_transform(X_train)
y_scaled = sc_y.fit_transform(y_train.reshape(-1,1))
```

```python
from sklearn.svm import SVR
svr = SVR(kernel='rbf', gamma='scale')
svr.fit(X_scaled, y_scaled.ravel())
```

```python
cv_svr = cross_val_score(estimator=svr, X=X_scaled, y=y_scaled.ravel(), cv=10)

#Predicting r2 score on train set
y_train_svr = sc_y.inverse_transform(svr.predict(sc_X.transform(X_train)))
r2_score_svr_train = r2_score(y_train, y_train_svr)

#Predicting r2 score on test set
y_pred_svr = sc_y.inverse_transform(svr.predict(sc_X.transform(X_test)))
r2_score_svr_test = r2_score(y_test, y_pred_svr)

# Predicting RMSE on test result
rmse_svr = (np.sqrt(mean_squared_error(y_test, y_pred_svr)))

print("CV: ", round(cv_svr.mean(),2))
print("Train Score: ", round(r2_score_svr_train,2))
print("Test Score: ", round(r2_score_svr_test,2))
print("Test Score(RMSE): ", rmse_svr)
```

# Decision Tree Regression

```python
from sklearn.tree import DecisionTreeRegressor
dt_regressor = DecisionTreeRegressor(random_state=0)
dt_regressor.fit(X_train, y_train)


cv_dt_regressor = cross_val_score(estimator=dt_regressor, X=X_train, y=y_train, cv=10)

#Predicting r2 score on train set
y_train_dt_regressor = dt_regressor.predict(X_train)
r2_score_dt_regressor_train = r2_score(y_train, y_train_dt_regressor)

#Predicting r2 score on test set
y_pred_dt_regressor = dt_regressor.predict(X_test)
r2_score_dt_regressor_test = r2_score(y_test, y_pred_dt_regressor)

# Predicting RMSE on test result
rmse_dt_regressor = (np.sqrt(mean_squared_error(y_test, y_pred_dt_regressor)))

print("CV: ", round(cv_dt_regressor.mean(),2))
print("Train Score: ", round(r2_score_dt_regressor_train,2))
print("Test Score: ", round(r2_score_dt_regressor_test,2))
print("Test Score(RMSE): ", rmse_dt_regressor)
```

# Random Forest Regressor

```python
from sklearn.ensemble import RandomForestRegressor
rf_regressor = RandomForestRegressor(n_estimators=1200, random_state=0)
rf_regressor.fit(X_train, y_train.ravel())


cv_rf_regressor = cross_val_score(estimator=rf_regressor, X=X_scaled, y=y_train.ravel(), cv=10)

#Predicting r2 score on train set
y_train_rf_regressor = rf_regressor.predict(X_train)
r2_score_rf_regressor_train = r2_score(y_train, y_train_rf_regressor)

#Predicting r2 score on test set
y_pred_rf_regressor = rf_regressor.predict(X_test)
r2_score_rf_regressor_test = r2_score(y_test, y_pred_rf_regressor)

# Predicting RMSE on test result
rmse_rf_regressor = (np.sqrt(mean_squared_error(y_test, y_pred_rf_regressor)))

print("CV: ", round(cv_rf_regressor.mean(),2))
print("Train Score: ", round(r2_score_rf_regressor_train,2))
print("Test Score: ", round(r2_score_rf_regressor_test,2))
print("Test Score(RMSE): ", rmse_rf_regressor)
```

# Measuring the error

```python
models = [('Linear', train_score, test_score, rmse_linear, cv_linear.mean()),
          ('Ploynomial', r2_score_polyreg_train, r2_score_polyreg_test, rmse_linear_poly,cv_linear_poly2.mean()),
          ('Ridge', r2_score_ridge_train, r2_score_ridge_test, rmse_ridge, cv_ridge.mean()),
          ('Lasso', r2_score_lasso_train, r2_score_lasso_test, rmse_lasso, cv_lasso.mean()),
          ('SVR', r2_score_svr_train, r2_score_svr_test, rmse_svr, cv_svr.mean()),
          ('Decision Tree', r2_score_dt_regressor_train, r2_score_dt_regressor_test, rmse_dt_regressor, cv_dt_regressor.mean()),
          ('Random Forest', r2_score_rf_regressor_train, r2_score_rf_regressor_test, rmse_rf_regressor, cv_rf_regressor.mean())]
```
